```python
import numpy as np
import matplotlib.pyplot as plt

# Time: 24 hours split into 100 points
hours = np.linspace(0, 24, 100)

# Temperature as a sine wave: mean 20°C, swing ±5°C
temperature = 20 + 5 * np.sin((2 * np.pi / 24) * hours)

# Plot
plt.plot(hours, temperature)
plt.xlabel("Hour of the day")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature Simulation")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Time: 24 hours split into 100 points
hours = np.linspace(0, 24, 100)

# Outside temperature as a sine wave: mean 10°C, swing ±10°C
outside_temp = 10 + 10 * np.sin((2 * np.pi / 24) * hours)

# Comfort temperature inside
comfort_temp = 21  

# Heat loss factors (bigger = more energy needed)
bad_insulation = 1.5  
good_insulation = 0.5  

# Energy use calculation
energy_bad = (comfort_temp - outside_temp) * bad_insulation
energy_good = (comfort_temp - outside_temp) * good_insulation

# Clip at 0 (if outside is warmer than inside, no heating needed)
energy_bad = np.clip(energy_bad, 0, None)
energy_good = np.clip(energy_good, 0, None)

# Plot
plt.figure(figsize=(10,6))
plt.plot(hours, outside_temp, color="gray", linestyle="--", label="Outside Temp (°C)")
plt.plot(hours, energy_bad, color="red", label="Energy Use: Poor Insulation")
plt.plot(hours, energy_good, color="green", label="Energy Use: Good Insulation")

plt.xlabel("Hour of the Day")
plt.ylabel("Energy Demand (arbitrary units)")
plt.title("Daily Heating Demand: Poor vs Good Insulation")
plt.legend()
plt.grid(True)
plt.show()
```
